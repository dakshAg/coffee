```python
import polars as pl
import pandas as pd
import numpy as np
```

```python
coffee_futures = pl.read_csv("../../data/landing/coffee_futures.csv")
relevant_stocks = pl.read_csv("../../data/landing/relevant_stocks.csv")
weather_data = pl.read_csv("../../data/landing/weather_data.csv")
```

```python
# Join the two dataframes on the date column
joined_data = coffee_futures.join(relevant_stocks, on="Date", how="inner")
joined_data = joined_data.join(weather_data, left_on="Date", right_on="date", how="inner")
```
