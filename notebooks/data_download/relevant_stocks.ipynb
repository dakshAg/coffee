```python
import yfinance as yf
```

```python
dat = yf.Tickers("COFF.L SBUX SJM BROS")
data = dat.history(period="1y")
data.head()
```

```python
# Flatten the column MultiIndex to single header
data.columns = ["_".join(filter(None, col)).strip() for col in data.columns.values]
```

```python
data.to_csv("../../data/landing/relevant_stocks.csv")
```
