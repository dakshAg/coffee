```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
dat = yf.Ticker("KC=F")
data = dat.history(period="1y")
data.head()
```

```python
# Convert the index to datetime if it's not already, then format as yyyy-mm-dd
data.index = pd.to_datetime(data.index)
data.index = data.index.strftime("%Y-%m-%d")
```

```python
data.to_csv("../../data/landing/coffee_futures.csv")
```
