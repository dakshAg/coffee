```python
import pandas as pd
import seaborn as sns
import altair as alt
```

```python
df = pd.read_csv("../../data/landing/weather_data.csv")
df.head()
```

```python
# Simple Altair plot for temperature (Mean, Max, Min) over Date

# Ensure date is in datetime format
df['date'] = pd.to_datetime(df['date'])

temp_df = df.melt(id_vars='date', value_vars=['temperature_2m_mean', 'temperature_2m_max', 'temperature_2m_min'],
                  var_name='Temperature Type', value_name='Temperature')

alt.Chart(temp_df).mark_line().encode(
    x=alt.X('date:T', title='Date'),
    y=alt.Y('Temperature:Q', title='Temperature (°C)'),
    color=alt.Color('Temperature Type:N', title='Temperature Type'),
    tooltip=['date:T', 'Temperature Type:N', 'Temperature:Q']
).properties(
    title='Temperature (Mean, Max, Min) over Date',
    width=800,
    height=400
).interactive()
```

```python


# Ensure date is in datetime format (should already be, but safe to reconvert)
df['date'] = pd.to_datetime(df['date'])

precip_chart = alt.Chart(df).mark_line(color='teal').encode(
    x=alt.X('date:T', title='Date'),
    y=alt.Y('precipitation_sum:Q', title='Precipitation (mm)'),
    tooltip=['date:T', 'precipitation_sum:Q']
).properties(
    title='Precipitation Sum over Date',
    width=800,
    height=400
).interactive()

precip_chart
```

```python
# Altair plot for daylight duration over Date

# Ensure date is in datetime format (should already be, but safe to reconvert)
df['date'] = pd.to_datetime(df['date'])

daylight_chart = alt.Chart(df).mark_line(color='orange').encode(
    x=alt.X('date:T', title='Date'),
    y=alt.Y('daylight_duration:Q', title='Daylight Duration (seconds)'),
    tooltip=['date:T', 'daylight_duration:Q']
).properties(
    title='Daylight Duration over Date',
    width=800,
    height=400
).interactive()

daylight_chart
```
